```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns

%reload_ext autoreload
%autoreload 2
```

```python
df = pd.read_csv (r'C:\Users\Pichau\Desktop\vgsales.csv')
df.dropna(how = "any",inplace = True)
print (df)
```

```python
#Dados
consoles_year = df.pivot_table('Global_Sales', index='Year', columns='Platform', aggfunc=sum)
main_consoles = consoles_year[['X360', 'PS', 'Wii', 'WiiU', 'PS2', 'PS3']]
#Grfico
main_consoles.plot(subplots=True, figsize=(16, 16), grid=False, title="Global sales per year")
```

Observando o grfico acima vemos que os consoles como PS2, PS3, e Xbox360 permanecem ativos por mais tempo no mercado que o Wii ou WiiU. Esses consoles apresentam um maior nmero de vendas de jogos deles.

# Como  a relao entre Gneros dos Jogos e as regies de vendas?

```python
var = df.groupby(['Genre']).sum()
var
```

 possvel perceber com essa tabela que existe uma enorme diferena na popularidade dos genros da industria entre as diversas regies do globo.  

```python
#Organizando Dados
df2 = df[df.columns.difference(['Rank', 'Year', 'Global_Sales'])]

var = df2.groupby(['Genre']).sum().sort_values('NA_Sales', ascending=True)
#Plotando Grfico
var.plot(kind='barh', 
                   stacked=True, 
                   color=['salmon','royalblue','gold','forestgreen','pink','crimson','cyan','slategray','midnightblue','darkred','burlywood','darkolivegreen'], 
                   grid=False, 
                   figsize=(15,8),title="Vendas de genros por regio")
```

Esse grfico pe em evidncia a diferena da popularidade dos genros entre as diferentes regies analsadas. Se observarmos com ateno fica claro uma prefrencia por jogos de Ao e Esporte nos Estados Unidos e Europa, e por jogos de Role_playing no Japo. Entretanto como evoluiram as vendas atravs dos anos?

# Como se sobresaem ttulos de franquias grandes?

```python
#Organizando dados
Gta_titles = df.pivot_table('Global_Sales', index='Year', columns='Name', aggfunc=sum)
Gta_titles
GTA_console = Gta_titles[['Grand Theft Auto V', 'Grand Theft Auto: Chinatown Wars', 'Grand Theft Auto IV', 'Grand Theft Auto: San Andreas', 'Grand Theft Auto: Vice City', 'Grand Theft Auto: Liberty City Stories', 'Grand Theft Auto III']]
GTA_console
data = {'Grand Theft Auto V': 55.92,
        'Grand Theft Auto: Chinatown Wars': 2.43,
        'Grand Theft Auto IV': 22.47,
        'Grand Theft Auto: San Andreas': 23.86,
        'Grand Theft Auto: Liberty City Stories': 11.26,
        'Grand Theft Auto: Vice City': 16.19,
        'Grand Theft Auto III': 13.11,}
group_data = list(data.values())
group_names = list(data.keys())
group_mean = np.mean(group_data)
#Plotando o grfico
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set(xlim=[0, 60], xlabel='Global Sales', ylabel='Jogos ordenados por lanamento',
       title='Franquia GTA')
```

Observando o grfico acima vemos que a franquia de ao Grand theft auto, cresce progressivamente a cada lanamento com excesso dos ttulo Liberty City Stories, ChinaTown Wars - jogos spin-off para consoles portteis -, e Grand Theft Auto IV, que no  maior que o predecessor por uma pequena margem.
Entretanto isso  vlido para jogos de todos os genros.

```python
poki = df[df.Name.str.startswith('Pokemon')].sort_values('Year', ascending=True)
poki_top5 = poki[['Name', 'Global_Sales']].head(15).set_index('Name')
poki_top5.plot(kind='bar', figsize=(8,8), grid=False, color='royalblue', legend=False).get_xticklabels()
plt.title('Os jogos mais vendidos da franquia Pokemon') 
plt.xlabel('Total de vendas') 
plt.ylabel('')
plt.show
```

Observando o grfico a seguir vemos que a saga Pokemon se sobresai de forma oposta a Grand Thef Auto. Se focarmos somente nos lanamentos principais da saga, Pokemon Red/Blue, Gold/Silver, e Ruby/Saphire, se torna evidente que a cada lanamento a franquia vende menos.
